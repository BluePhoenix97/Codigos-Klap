K Means

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Asumiendo que el archivo se llama 'CadenasClusters.xlsx'
data = pd.read_excel('CadenasClusters.xlsx')

# Convertir la columna 'MA%' a porcentaje
data['MA%'] = data['MA%'] * 100

# Seleccionar las columnas relevantes para el clustering
# Aquí se eligen 'Facturación', 'MA' y 'MA%' como ejemplo
X = data[['Facturación', 'MA', 'MA%']]

# Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Inicializar el modelo K-Means para 3 clústeres
kmeans = KMeans(n_clusters=3, random_state=0)

# Aplicar el modelo K-Means
data['Cluster'] = kmeans.fit_predict(X_scaled)

# Visualizar los primeros resultados
print(data.head())

# Visualizar los clusters
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=data['Cluster'], cmap='viridis')
plt.xlabel('Facturación (estandarizada)')
plt.ylabel('MA% (estandarizado)')
plt.title('Clusters de K-Means')
plt.show()
```

```python
# Contar el número de elementos en cada cluster
cluster_counts = data['Cluster'].value_counts()
print(cluster_counts)
```

```python
# Obtener los nombres de cada cadena en cada cluster
# Suponiendo que tienes una columna 'Cadena' que contiene los nombres
for cluster in sorted(data['Cluster'].unique()):
    print(f"Cluster {cluster}:")
    cadenas = data[data['Cluster'] == cluster]['cadena'].tolist()
    print(cadenas)
    print()
```
